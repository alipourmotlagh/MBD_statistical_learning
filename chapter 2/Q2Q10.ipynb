```r
# B
#install.packages('ggplot2')
#install.packages('ggpubr')
#install.packages('ggthemes')
library(ggplot2)
library(tidyverse)
library(ggpubr)
library(ggthemes)
```

```r
pairs(crim ~ ., data=Boston)
```

```r
#some other paair wise plots with crime
g1<-Boston%>%ggplot(aes(age,crim))+geom_point()+xlab("proportion of owner-occupied units built prior to 1940")+ylab("Crime")
g2<-Boston%>%ggplot(aes(zn,crim))+geom_point()+xlab("proportion of residential land zoned for lots over 25,000 sq.ft.")+ylab("Crime")
g2<-Boston%>%ggplot(aes(indus,crim))+geom_point()+xlab("proportion of non-retail business acres per town")+ylab("Crime")
g3<-Boston%>%ggplot(aes(chas,crim))+geom_point()+xlab("Charles River ")+ylab("Crime")
g4<-Boston%>%ggplot(aes(nox,crim))+geom_point()+xlab("nitric oxides concentration (parts per 10 million)")+ylab("Crime")
g5<-Boston%>%ggplot(aes(rm,crim))+geom_point()+xlab("average number of rooms per dwelling")+ylab("Crime")
g6<-Boston%>%ggplot(aes(dis,crim))+geom_point()+xlab("weighted distances to five Boston employment centres")+ylab("Crime")

ggarrange(g1, g2,g3,g4,g5,g6, ncol=2,nrow=3)
```

```r
#part2
g7<-Boston%>%ggplot(aes(rad,crim))+geom_point()+xlab("index of accessibility to radial highways")+ylab("Crime")
g8<-Boston%>%ggplot(aes(tax,crim))+geom_point()+xlab("full-value property-tax rate per 10,000")+ylab("Crime")
g9<-Boston%>%ggplot(aes(ptratio,crim))+geom_point()+xlab("pupil-teacher ratio by town")+ylab("Crime")
g10<-Boston%>%ggplot(aes(black,crim))+geom_point()+xlab("1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town ")
+ylab("Crime")
g11<-Boston%>%ggplot(aes(lstat,crim))+geom_point()+xlab("percentage lower status of the population")+ylab("Crime")
g12<-Boston%>%ggplot(aes(medv,crim))+geom_point()+xlab("Median value of owner-occupied homes in 1000's ")+ylab("Crime")

ggarrange(g7, g8,g9,g10,g11,g12, ncol=2,nrow=3)
```

```r
# C
for (x in colnames(Boston)) {
    if (!(x %in% c('crim'))) {
        rel <- cor.test(Boston$crim, Boston[, x], method="pearson")
        print(x)
        print(paste('corelation =', rel$estimate))
        print(paste('pvalue =', rel$p.value))
    }
}
```
